# Word Cloud

# Simple Word Cloud using "Lord of the Rings" book stored in github

##This code combines three different list of stopwords

```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
from nltk.corpus import stopwords
```

```python
import matplotlib.pyplot as plt
```

# process the file

```python

import requests

# Replace the URL below with the raw URL of your text file on GitHub
url = 'https://raw.githubusercontent.com/ganesh-k13/shell/refs/heads/master/test_search/www.glozman.com/TextPages/01%20-%20The%20Fellowship%20Of%20The%20Ring.txt'

response = requests.get(url)
response.raise_for_status()  # Raise an error for HTTP issues
content = response.text
content = content.lower()
text = content
```

```python

# Stop words list #1

# Create stopword list:
wordcloud_stopwords = set(STOPWORDS)

# Download NLTK stopwords if not already present
nltk.download("stopwords")

# Stop word list #2

# Get the NLTK stopwords
nltk_stopwords = set(stopwords.words("english"))
# Combine both sets of stopwords
combined_stopwords = wordcloud_stopwords.union(nltk_stopwords)

# Stop words list #3

stopwords_list = requests.get("https://gist.githubusercontent.com/rg089/35e00abf8941d72d419224cfd5b5925d/raw/12d899b70156fd0041fa9778d657330b024b959c/stopwords.txt").content
combined_stopwords = set(stopwords_list.decode().splitlines())

# text = "We planted flowers in the garden. He sent her a bouquet of flowers. He wore a single flower in his lapel"


# Generate a word cloud image
wordcloud = WordCloud(stopwords=combined_stopwords, background_color="white").generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
